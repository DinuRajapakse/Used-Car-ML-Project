# ML Model For Predicting Used Car Selling Prices. 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import activations
import csv
import keras_tuner as kt
from keras_tuner.tuners import Hyperband
from keras_tuner.tuners import RandomSearch

import pickle
```

```python
df =pd.read_csv('car data.csv')
df
```

```python
final_df = pd.get_dummies(df, columns = ['Fuel_Type', 'Seller_Type','Transmission'], prefix = ['','',''])
```

```python
final_df.drop('Car_Name',inplace=True, axis=1)
final_df = final_df.astype(np.float64)
final_df
```

```python
# final_df = final_df.reset_index(drop = True)

train_dataset = final_df.sample(frac=0.8, random_state=13)
test_dataset = final_df.drop(train_dataset.index)

y_train_k = train_dataset.pop('Selling_Price')
y_test_k = test_dataset.pop('Selling_Price')

train_stats = train_dataset.describe()
train_stats = train_stats.transpose()

test_stats = test_dataset.describe()
test_stats = test_stats.transpose()

train_stats
```

```python
def norm(x):
    return(x-train_stats['mean']) / train_stats['std'] #standardize
#     return(x-train_stats['min']) / (train_stats['max'] -  train_stats['min']) #normalize
normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)


# normed_train_data = train_dataset
# normed_test_data = test_dataset
```

```python
def build_model(tuner):
    model = keras.models.Sequential()
    model.add(layers.Dense(tuner.Int(min_value=3,max_value=9, step=2, name='Dense_1'),activation = activations.relu,input_shape=[len(train_dataset.keys())]))
    model.add(layers.Dense(tuner.Int(min_value=3,max_value=9, step=2, name='Dense_2'),activation = activations.relu))
    model.add(layers.Dense(tuner.Int(min_value=3,max_value=9, step=2, name='Dense_2'),activation = activations.relu))
    model.add(layers.Dense(tuner.Int(min_value=3,max_value=9, step=2, name='Dense_2'),activation = activations.relu))
    
    for i in range(tuner.Int('#of_layers', 1, 3)):
        model.add(layers.Dense(tuner.Int(min_value=2,max_value=9,step=2,name=f"{i+1}th layer"), activation = activations.relu))
        
    model.add(layers.Dense(1))
    
    optimizer = keras.optimizers.Adam(learning_rate=0.0001)
    model.compile(loss='mse',
                  optimizer = optimizer,
                  metrics = ['mae', 'mse'])   
    
    return model
```

```python
class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100==0:
            print('.',end='')
EPOCHS=1000
```

```python
print("Instantiating a random search tuner object...")
tuner = kt.Hyperband(build_model, objective = 'mae', max_epochs=EPOCHS, seed=42, overwrite = True)
# tuner = kt.RandomSearch(build_model, objective = 'val_accuracy', max_trials = 3, executions_per_trial = 5)
```

```python
print("[INFO] performing hyperparameter search...")
print(tuner.search(normed_train_data.to_numpy(), y_train_k.to_numpy(), callbacks = [PrintDot()], epochs=EPOCHS))
```

```python
# with open(f'tuner_{int(time.time())}.pkl', wb) as f:
#     pickle.dump(tuner, f)

tuner.results_summary(1)
```

```python
print(tuner.get_best_models()[0].summary())
```

```python
history = tuner.get_best_models()[0].fit(
    normed_train_data, y_train_k,
    epochs=EPOCHS, validation_split = 0.2, verbose=0,
    callbacks=[PrintDot()])

EPOCHS=1000

hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
import matplotlib.pyplot as plt

def plot_history(history):
    hist = pd.DataFrame(history.history)
    hist['epoch']=history.epoch
    
    plt.figure()
    plt.xlabel('Epoch')
    plt.ylabel('Mean Abs Error [price]')
    plt.plot(hist['epoch'], hist['mae'],
             label = 'Train Error')
    plt.plot(hist['epoch'], hist['val_mae'],
             label = 'Val Error')
    plt.legend()
    plt.ylim([0,1])
    
    plt.figure()
    plt.xlabel('Epoch')
    plt.ylabel('Mean Square Error [$price^2$]')
    plt.plot(hist['epoch'], hist['mse'],
            label='Train Error')
    plt.plot(hist['epoch'], hist['val_mse'],
            label = 'Val Error')
    plt.legend()
    plt.ylim([0,2.5])
    
plot_history(history)

# model = build_model()

# loss, mae, mse = model.evaluate(normed_test_data, y_test_k, verbose=0)
```

```python
loss, mae, mse = tuner.get_best_models()[0].evaluate(normed_test_data, y_test_k, verbose=0)
test_predictions = tuner.get_best_models()[0].predict(normed_test_data).flatten()

plt.scatter(y_test_k, test_predictions)
plt.xlabel('True Values [price]')
plt.ylabel('Predictions [price]')
# plt.axis('equal')
# plt.axis('square')
# plt.xlim([-10, plt.xlim()[3]])
# plt.ylim([-10, plt.ylim()[3]])
_ = plt.plot([0, 25], [0, 25])

print('Testing set Mean Abs Error: {:5.2f}'.format(mae))
print('Testing set Mean Squared Error: {:5.2f}'.format(mse))
print('Testing set loss: {:5.2f}'.format(loss))
```
